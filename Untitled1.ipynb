```python
def extract_mfcc(file_path):
    # تحميل ملف الصوت
    audio, sample_rate = librosa.load(file_path, duration=3, offset=0.5)

    # استخراج الـ MFCCs
    mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)

    # حساب المتوسط عشان نوحد حجم الميزة (Feature Scaling)
    mfccs_processed = np.mean(mfccs.T, axis=0)
    return mfccs_processed

# مثال: لو عندك ملف اسمه 'speech.wav'
# features = extract_mfcc('speech.wav')
# print(features.shape)
```

```python
import librosa
import numpy as np
import glob

# دالة لاستخراج الـ Features
def get_features(file_path):
    # تحميل الصوت
    audio, sr = librosa.load(file_path, duration=3, offset=0.5)
    # استخراج الـ MFCCs
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=40)
    # نأخذ المتوسط لتوحيد الحجم
    return np.mean(mfccs.T, axis=0)

X = []
y = []

# البحث عن كل ملفات الـ wav جوه الفولدرات
files = glob.glob("/content/Actor_*/*.wav")

for file in files:
    # الـ Label موجود في الجزء التالت من اسم الملف (مثلاً 03-01-03-...)
    # 03 هنا يمثل Happy
    label = int(file.split("-")[2])
    features = get_features(file)

    X.append(features)
    y.append(label)

X = np.array(X)
y = np.array(y)
print("Data shape:", X.shape)
```

```python
import librosa
import numpy as np

def extract_spectrogram(file_path):
    audio, sr = librosa.load(file_path, duration=3, offset=0.5)
    # تحويل الصوت لـ Mel-Spectrogram
    spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=64)
    # تحويل لـ Log-scale
    spectrogram = librosa.power_to_db(spectrogram, ref=np.max)
    return spectrogram

# تجميع الـ Spectrograms
X_cnn = []
y_labels = []

# (تأكد إن مسار الملفات صحيح زي ما ظهر عندك في القائمة)
for file in files:
    spec = extract_spectrogram(file)
    # CNN بيحتاج أبعاد ثابتة، هنخليها (64, 130)
    if spec.shape[1] == 130:
        X_cnn.append(spec)
        label = int(file.split("-")[2]) - 1 # تعديل الـ Label ليبدأ من 0
        y_labels.append(label)

X_cnn = np.array(X_cnn)
X_cnn = X_cnn[..., np.newaxis] # إضافة بعد إضافي للـ CNN
y_labels = np.array(y_labels)

print("Data shape for CNN:", X_cnn.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 130, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(8, activation='softmax') # 8 مشاعر في RAVDESS
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# التدريب
model.fit(X_cnn, y_labels, epochs=10, batch_size=32)
```
